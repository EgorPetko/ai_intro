```python
#8
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

d = pd.read_csv("student_data.csv", sep=',')
params = ['failures','Medu','Fedu','goout','studytime'] #['failures','Medu','Fedu','goout','studytime']
sh = 0
col = ['blue','green','black','yellow','red']
for to in params:
  st = "G3"
  map1 = dict()
  map2 = dict()
  y_ = []
  x_ = []
  i = 0
  while(i < len(d[st])):
    if d[st][i] in map1:
      map1[data[st][i]] += d[to][i]
      map2[data[st][i]] += 1
    else :
      map1[data[st][i]] = d[to][i]
      map2[data[st][i]] = 1
    i += 1
  for j in map1:
    x_.append(j)
    y_.append(map1[j]/map2[j])

  correlation = np.corrcoef(data[st], data[to])[0, 1]
  print(" :", correlation)
  sns.regplot(x=st, y=to, data=d);
  sns.regplot(x = x_, y = y_)
  plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
data = pd.read_csv("student_data.csv", sep=',')
data.describe()

for name in data:
  if(data[name].dtypes == data["G3"].dtypes) : continue
  map = dict()
  sh = 0
  for to in data[name]:
    if(to in map) : continue
    map[to] = sh
    sh += 1
  data[name].replace(map, inplace=True)

for name2 in data:
  print(np.corrcoef(data["G1"], data[name2])[0, 1] * 100,name2)
```
